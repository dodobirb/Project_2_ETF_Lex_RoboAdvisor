# Daily & Total Returns

```python
spdr_etfs_df = pd.concat([
    technology_xlk_df,
    healthcare_xlv_df,
    financials_xlf_df,
    realestate_xlre_df,
    energy_xle_df,
    materials_xlb_df,
    consumerdisc_xly_df,
    industrial_xli_df,
    utilities_xlu_df,
    consumerstaples_xlp_df,
    communications_xlc_df], axis=1)

spdr_etfs_df
```

```python
spdr_etfs_df.plot(figsize=(20, 8), title="S&P Sector ETFs --- Jan. 2011 - Jan. 2021")

print("Note: Real Estate XLRE & Communications XLC were created after the start of the overall dataset.")
```

```python
daily_returns_spdr_etfs = spdr_etfs_df.pct_change()

daily_returns_spdr_etfs = daily_returns_spdr_etfs.iloc[1::]

daily_returns_spdr_etfs
```

```python
total_returns_spdr_etfs = pd.DataFrame(daily_returns_spdr_etfs.sum() * 100).sort_values(0)
total_returns_spdr_etfs.columns=['% Change']

total_returns_spdr_etfs
```

```python
total_returns_spdr_etfs.hvplot.bar(
    title="ETF Total Returns --- Jan. 2011 - Jan. 2021",
    width=1000, 
    height=500,
    legend=False, 
    rot=45
).opts(
    color='% Change')
```

# Volatility Visualized

```python
daily_returns_spdr_etfs.hvplot.box(
    title="ETF Daily Volatility --- Jan. 2011 - Jan. 2021",
    width=1000, 
    height=500,
    rot=45)
```

# Risk - By the Numbers & Visualized

```python
stdev_etfs = spdr_etfs_df.std().sort_values()

stdev_etfs
```

```python
annualized_stdev_etfs = stdev_etfs * np.sqrt(252)

annualized_stdev_etfs.sort_values()
```

```python
rolling_stdev_etfs = daily_returns_spdr_etfs.rolling(window=90).std()

rolling_stdev_etfs.plot(
    figsize=(20, 10),
    title="ETFs 3-Month Rolling Standard Deviation --- Jan. '11 - Jan. '21"
)
```

# Risk-Return Profile

```python
annualized_avg_returns = daily_returns_spdr_etfs.mean() * 252

annualized_avg_returns.sort_values()
```

```python
annualized_sharpe_ratios = (annualized_avg_returns / annualized_stdev_etfs).sort_values()
```

```python
annualized_sharpe_ratios.plot.bar(
    title="ETFs Annualized Sharpe Ratios --- Jan. '11 - Jan. '21",
    figsize=(20, 8),
    legend=False,
    ylabel="Sharpe Ratio",
    rot=45
)
```

Upload results to S3 for use in Lex application

```python
### Start code to build JSON file with ETF data uploaded to S3 bucket used in Lambda and Lex application

import json

#List provides long name to use in JSON file sent to S3 bucket
etf_list = ['Consumer Discretionary Select Sector SPDR','Consumer Staples Select Sector SPDR','Communication Services Select Sector SPDR',
            'Energy Select Sector SPDR','Financial Select Sector SPDR','Health Care Select Sector SPDR','Industrial Select Sector SPDR',
            'Materials Select Sector SPDR','Real Estate Select Sector SPDR','Technology Select Sector SPDR', 'Utilities Select Sector SPDR']

#Combines multiple risk/return aggregated data series into a single dataframe
combined_df = pd.DataFrame({'Annualized_StDev':annualized_stdev_etfs, 'Annualized_Sharpe_Ratio':annualized_sharpe_ratios, 'Annualized_AvgReturns':annualized_avg_returns})

#Joins the total returns dataframe to make a combined dataframe
combined_df = combined_df.join(total_returns_spdr_etfs)

#Adding enriched data fields to the dataframe to identify each ETF along with it's risk level
combined_df["Ticker"] = ["XLY","XLP","XLC","XLE","XLF","XLV","XLI","XLB","XLRE","XLK","XLU"]

combined_df["Name"] = etf_list

combined_df["Risk_Level"] = ["High","Medium","Low","Medium","Low","High","Medium","Medium","None","High","Medium"]

#Transposing the dataframe and adding ticker as columns to make it easier to utilize in Lex - ticker as the key in the dictionary
#makes it simplier to identify the correct ETF based on the users selection
combined_df = combined_df.transpose()

combined_df.columns = ["XLY","XLP","XLC","XLE","XLF","XLV","XLI","XLB","XLRE","XLK","XLU"]

output = combined_df.to_dict()

#output to screen and to local drive to preparer for upload via boto3 client. Please update with your own file location
#path and do not delete 'r' or python will think some of the path are escape keys
with open(r'C:\Users\bmill\GIT\Project_2\Project_2_ETF_Lex_RoboAdvisor\Resources\returns.json', 'w') as f:
    json.dump(output, f)

print(output)
```
