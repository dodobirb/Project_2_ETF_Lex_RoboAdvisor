```python
import investpy.etfs as etfs
import pandas as pd
```

```python
technology_xlk_df = etfs.get_etf_historical_data(etf="Technology Select Sector SPDR", country="united states", from_date="04/01/2011", to_date="04/01/2021")
technology_xlk_df = technology_xlk_df.drop(columns=["Open", "High", "Low", "Volume", "Currency", "Exchange"]).rename(columns={'Close': 'Close Technology XLK'})

technology_xlk_df
```

```python
spdr_etfs_df = pd.concat([
    technology_xlk_df,
    healthcare_xlv_df,
    financials_xlf_df,
    realestate_xlre_df,
    energy_xle_df,
    materials_xlb_df,
    consumerdisc_xly_df,
    industrial_xli_df,
    utilities_xlu_df,
    consumerstaples_xlp_df,
    communications_xlc_df], axis=1)

spdr_etfs_df
```

```python
print("Note: Real Estate XLRE & Communications XLC were created after the start of the overall dataset.")
spdr_etfs_df.plot(figsize=(20, 8))
```

```python
daily_returns_spdr_etfs = spdr_etfs_df.pct_change().drop('2011-01-04')

daily_returns_spdr_etfs
```

```python
total_returns_spdr_etfs = pd.DataFrame(daily_returns_spdr_etfs.sum()).transpose()

total_returns_spdr_etfs
```
