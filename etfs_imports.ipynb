# Daily & Total Returns

```python
spdr_etfs_df = pd.concat([
    technology_xlk_df,
    healthcare_xlv_df,
    financials_xlf_df,
    realestate_xlre_df,
    energy_xle_df,
    materials_xlb_df,
    consumerdisc_xly_df,
    industrial_xli_df,
    utilities_xlu_df,
    consumerstaples_xlp_df,
    communications_xlc_df], axis=1)

spdr_etfs_df
```

```python
spdr_etfs_df.plot(figsize=(20, 8), title="S&P Sector ETFs --- Jan. 2011 - Jan. 2021")

print("Note: Real Estate XLRE & Communications XLC were created after the start of the overall dataset.")
```

```python
daily_returns_spdr_etfs = spdr_etfs_df.pct_change().drop('2011-01-04')

daily_returns_spdr_etfs
```

```python
total_returns_spdr_etfs = pd.DataFrame(daily_returns_spdr_etfs.sum() * 100).sort_values(0)
total_returns_spdr_etfs.columns=['% Change']

total_returns_spdr_etfs
```

```python
total_returns_spdr_etfs.hvplot.bar(
    title="ETF Total Returns --- Jan. 2011 - Jan. 2021",
    width=1000, 
    height=500,
    legend=False, 
    rot=45
).opts(
    color='% Change')
```

# Volatility Visualized

```python
daily_returns_spdr_etfs.hvplot.box(
    title="ETF Daily Volatility --- Jan. 2011 - Jan. 2021",
    width=1000, 
    height=500,
    rot=45)
```

# Risk - By the Numbers & Visualized

```python
stdev_etfs = spdr_etfs_df.std().sort_values()

stdev_etfs
```

```python
annualized_stdev_etfs = stdev_etfs * np.sqrt(252)

annualized_stdev_etfs.sort_values()
```

```python
rolling_stdev_etfs = daily_returns_spdr_etfs.rolling(window=90).std()

rolling_stdev_etfs.plot(
    figsize=(20, 10),
    title="ETFs 3-Month Rolling Standard Deviation --- Jan. '11 - Jan. '21"
)
```

# Risk-Return Profile

```python
annualized_avg_returns = daily_returns_spdr_etfs.mean() * 252

annualized_avg_returns.sort_values()
```

```python
annualized_sharpe_ratios = (annualized_avg_returns / annualized_stdev_etfs).sort_values()

annualized_sharpe_ratios
```

```python
annualized_sharpe_ratios.plot.bar(
    title="ETFs Annualized Sharpe Ratios --- Jan. '11 - Jan. '21",
    figsize=(20, 8),
    legend=False,
    ylabel="Sharpe Ratio",
    rot=45
)
```
